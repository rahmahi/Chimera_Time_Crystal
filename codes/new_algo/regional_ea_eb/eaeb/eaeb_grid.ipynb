```python
import matplotlib.pyplot as plt
import numpy as np

# Update the default parameters
plt.rcParams.update({"figure.figsize": (5, 6), "text.usetex": True, "font.family": "sans-serif",
                     "font.size": 12, "font.sans-serif": ["Helvetica"]})

# Data
ea = np.array([0.03, 0.03, 0.03, 0.03, 0.1, 0.3, 0.4, 0.5, 0.7, 0.8])
eb = np.array([0.1, 0.3, 0.7, 0.9, 0.11, 0.31, 0.1, 0.51, 0.1, 0.81])
x = np.linspace(0, 1, 10)
y = x

# Create subplots
fig,  ax_main = plt.subplots()
#plt.subplots_adjust(wspace=0.02, hspace=0.2)
ax_left = fig.add_axes([-0.03, 0.192, 0.05, 0.685])
ax_bottom = fig.add_axes([0.175, -0.03, 0.55,  0.05])

# Main plot
ax_main.plot(x, y, color='blue', alpha=0.2, linestyle='--', label=r'$\epsilon_A = \epsilon_B$')
ax_main.scatter(ea, eb, color='red')
ax_main.set_ylabel(r"$\epsilon_{_B}$", fontsize = 20, labelpad = 0)
ax_main.set_xlabel(r"$\epsilon_{_A}$", fontsize = 20, labelpad = 0)
ax_main.tick_params(which='both', axis='both', direction="in")
ax_main.legend(loc=7, frameon=True, fontsize=18)

# Left subplot
ax_left.scatter(np.ones(4)*0.03, np.flip(ea[0:4]-eb[0:4]), color='blue') 
ax_left.set_xlabel(r"$\epsilon_A$", labelpad = 0, fontsize=16)
ax_left.set_ylabel(r"$\epsilon_A-\epsilon_B$", labelpad = 0, fontsize=16)
ax_left.yaxis.set_label_position('left')
ax_left.set_xlim([-0.05, 0.08])
ax_left.set_xticks([0.03])
ax_left.set_yticks(np.flip(ea[0:4]-eb[0:4]))
ax_left.tick_params(axis='both', direction='in')
ax_left.invert_yaxis() 

# Bottom subplot
bottom_xdata = np.array([0.03, 0.4, 0.7])
bottom_ydata = np.ones(3) * 0.1
ax_bottom.scatter(bottom_xdata-bottom_ydata, bottom_ydata, color='green') 
ax_bottom.set_ylabel(r"$\epsilon_B$", labelpad = 10, fontsize=16, rotation = 0)
ax_bottom.set_xlabel(r"$\epsilon_A-\epsilon_B$", labelpad = 0, fontsize=16)
ax_bottom.yaxis.set_label_position('left')
ax_bottom.set_ylim(-0.15, 0.3)
ax_bottom.set_yticks([0.1])
ax_bottom.set_xticks(bottom_xdata-bottom_ydata)
ax_bottom.tick_params(axis='both', direction='in')

# Adjust layout
plt.tight_layout()

# Save and show
plt.savefig('eaeb_grid_with_subplots_right_bottom.svg', bbox_inches='tight', pad_inches=0.0, dpi=600)
plt.show()
```

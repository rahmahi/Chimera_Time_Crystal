```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.transforms as transforms

plt.rcParams.update({"figure.figsize": (7.5, 4.4),"text.usetex": True,"font.family": "sans-serif",
    "font.size": 12,"font.sans-serif": ["Helvetica"]})

ea = [0.03,0.03,0.03,0.03,0.1,0.3,0.4,0.5,0.7,0.8]
eb = [0.1,0.3,0.7,0.9,0.11,0.31,0.1,0.51,0.1,0.81]

x = np.linspace(0, 1, 10)
y = x

fig, ax1 = plt.subplots()
ax1.scatter(ea, eb, color='red')
ax1.plot(x, y, color='black', alpha=0.2, linestyle='--', label=r'$\epsilon_A = \epsilon_B$')
ax1.set_ylabel(r"$\epsilon_{_B}$", fontsize=16)
ax1.set_xlabel(r"$\epsilon_{_A}$", fontsize=16)
ax1.tick_params(which='both', axis='both', direction="in")
ax1.legend(loc=7, frameon=False, fontsize=14)

plt.show()
```

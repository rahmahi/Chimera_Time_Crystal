```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({"figure.figsize": (7.5, 4.4),"text.usetex": True,"font.family": "sans-serif",
    "font.size": 12,"font.sans-serif": ["Helvetica"]})

ea = [0.03,0.03,0.03,0.03,0.1,0.3,0.4,0.5,0.7,0.8]
eb = [0.1,0.3,0.7,0.9,0.11,0.31,0.1,0.51,0.1,0.81]

x = np.linspace(0, 1, 10)
y = x

plt.scatter(ea, eb, color = 'red')
plt.plot(x,y, color = 'black', alpha = 0.2, linestyle = '--', label = r'$\epsilon_A \approx \epsilon_B$')

plt.xlabel(r"$\epsilon_{_A}$", fontsize = 16)
plt.ylabel(r"$\epsilon_{_B}$", fontsize = 16)
plt.tick_params(which='both', axis="x", direction="in")
plt.tick_params(which='both', axis="y", direction="in")

plt.legend(loc = 7, frameon = False, fontsize = 14)

#plt.savefig('eaeb_grid.svg', bbox_inches='tight', pad_inches=0.0, dpi = 300)
plt.show()
```

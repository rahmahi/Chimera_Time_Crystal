```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.transforms as transforms

plt.rcParams.update({"figure.figsize": (4,6),"text.usetex": True,"font.family": "sans-serif",
    "font.size": 12,"font.sans-serif": ["Helvetica"]})

ea = [0.03,0.03,0.03,0.03,0.1,0.3,0.4,0.5,0.7,0.8]
eb = [0.1,0.3,0.7,0.9,0.11,0.31,0.1,0.51,0.1,0.81]

x = np.linspace(0, 1, 10)
y = x

plt.plot(x, y, color='blue', alpha=0.2, linestyle='--', label=r'$\epsilon_A = \epsilon_B$')
plt.scatter(ea, eb, color='red')
plt.ylabel(r"$\epsilon_{_B}$", fontsize=16)
plt.xlabel(r"$\epsilon_{_A}$", fontsize=16)
plt.tick_params(which='both', axis='both', direction="in")
plt.legend(loc=7, frameon=False, fontsize=14)

plt.savefig('eaeb_grid.svg', bbox_inches='tight', pad_inches=0.0, dpi = 600)

plt.show()
```

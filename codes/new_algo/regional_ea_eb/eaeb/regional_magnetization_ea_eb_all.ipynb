# Regional magnetization

\begin{equation}
M^z_{A/B} = \frac{2}{N} \sum_{l\in {A,B}}\langle \sigma^z_l (t)\rangle
\end{equation}

```python
!ls *_ea_0.03_eb_0.1_.hdf5>mzfiles_ea0.03_eb0.1.txt
!cat mzfiles_ea0.03_eb0.1.txt




import matplotlib.pyplot as plt
import h5py
import numpy as np

plt.rcParams.update({"figure.figsize": (8., 6),"text.usetex": True,"font.family": "sans-serif",
    "font.size": 16,"font.sans-serif": ["Helvetica"]})
w = 20
T = 2 * np.pi/w
jvalues = [0.072/T,0.2/T]

fig, axs = plt.subplots(4, 2, sharey=True, sharex=True)
plt.subplots_adjust(wspace=0.02, hspace=0.1)

with open("mzfiles_ea0.03_eb0.1.txt", 'r') as fp:
    lines = fp.readlines()
    x = len(lines)
    for i in range(0,x):
        filename = lines[i][:-1]
        print(filename)
        with h5py.File(filename, "r") as f:
            data_a = f['mza'][()]
            data_b = f['mzb'][()]
            times = f['times'][()]
            N = f.attrs['N']
            Jvalue = f.attrs['Jvalue']
            beta = f.attrs['beta']
            w = f.attrs['w']
            ea = f.attrs['ea']
            eb = f.attrs['eb']

            #print('beta',beta)
            
            if Jvalue == jvalues[0]:
                jh = 0
            else:
                jh = 1
                
            if beta == 0:
                ii = 0
                lbl = fr"$\beta={beta}$"
            if beta == 1.5:
                ii = 1
                lbl = fr"$\beta={beta}$"
            if beta == 2.5:
                ii = 2
                lbl = fr"$\beta={beta}$"
            if beta == float('inf'):
                ii = 3
                lbl = r"$\beta={\infty}$"
                
            
            axs[ii][jh].plot(times/T, data_a, color='red', label= r"$Region-A$")            
            axs[ii][jh].plot(times/T, data_b, color='blue', label= r"$Region-B$")
                
            axs[0][0].legend(frameon = True, loc = 6, fontsize = 12)

            axs[ii][jh].set_xlim(0, 80)
            axs[ii][jh].set_ylim(-1,1)
            #axs[hh][jh].set_xscale('log')
            axs[ii][jh].tick_params(which='both', axis="x", direction="in")
            axs[ii][jh].tick_params(which='both', axis="y", direction="in")
            axs[0][jh].set_title(fr"$\epsilon_A = {ea}, \epsilon_B = {eb}, J_0 = {Jvalue:1.3f}$", fontsize= 11)
            

            axs[ii][0].set_ylabel(r"$M^z_{A/B}$",fontsize = 14, labelpad= 0.0)  
            
            if ii ==3:
                axs[ii][1].text(20, 0.0, lbl, fontsize=12, bbox=dict(facecolor='white', alpha=1.0, edgecolor='white'))
            else:
                axs[ii][1].text(20, 0.0, lbl, fontsize=12, bbox=dict(facecolor='white', alpha=1.0, edgecolor='white'))
            
            
axs[3][0].set_xlabel(r"$t/T$", fontsize = 14, labelpad= 0.5)
axs[3][1].set_xlabel(r"$t/T$", fontsize = 14, labelpad= 0.5)

figname1 = 'reg_mz_ea_' + str(ea) + '_eb_' + str(eb) + '.svg'
figname2 = 'reg_mz_ea_' + str(ea) + '_eb_' + str(eb) + '.pdf'
plt.savefig(figname1, bbox_inches='tight', pad_inches=0.0, dpi = 300)
#plt.savefig(figname2, bbox_inches='tight', pad_inches=0.0, dpi = 600)
plt.show()
```

```python
!ls *_ea_0.4_eb_0.1_.hdf5>mzfiles_ea0.4_eb0.1.txt
!cat mzfiles_ea0.4_eb0.1.txt




import matplotlib.pyplot as plt
import h5py
import numpy as np

plt.rcParams.update({"figure.figsize": (8., 6),"text.usetex": True,"font.family": "sans-serif",
    "font.size": 16,"font.sans-serif": ["Helvetica"]})
w = 20
T = 2 * np.pi/w
jvalues = [0.072/T,0.2/T]

fig, axs = plt.subplots(4, 2, sharey=True, sharex=True)
plt.subplots_adjust(wspace=0.02, hspace=0.1)

with open("mzfiles_ea0.4_eb0.1.txt", 'r') as fp:
    lines = fp.readlines()
    x = len(lines)
    for i in range(0,x):
        filename = lines[i][:-1]
        print(filename)
        with h5py.File(filename, "r") as f:
            data_a = f['mza'][()]
            data_b = f['mzb'][()]
            times = f['times'][()]
            N = f.attrs['N']
            Jvalue = f.attrs['Jvalue']
            beta = f.attrs['beta']
            w = f.attrs['w']
            ea = f.attrs['ea']
            eb = f.attrs['eb']

            #print('beta',beta)
            
            if Jvalue == jvalues[0]:
                jh = 0
            else:
                jh = 1
                
            if beta == 0:
                ii = 0
                lbl = fr"$\beta={beta}$"
            if beta == 1.5:
                ii = 1
                lbl = fr"$\beta={beta}$"
            if beta == 2.5:
                ii = 2
                lbl = fr"$\beta={beta}$"
            if beta == float('inf'):
                ii = 3
                lbl = r"$\beta={\infty}$"
                
            
            axs[ii][jh].plot(times/T, data_a, color='red', label= r"$Region-A$")            
            axs[ii][jh].plot(times/T, data_b, color='blue', label= r"$Region-B$")
                
            axs[0][0].legend(frameon = True, loc = 6, fontsize = 12)

            axs[ii][jh].set_xlim(0, 80)
            axs[ii][jh].set_ylim(-1,1)
            #axs[hh][jh].set_xscale('log')
            axs[ii][jh].tick_params(which='both', axis="x", direction="in")
            axs[ii][jh].tick_params(which='both', axis="y", direction="in")
            axs[0][jh].set_title(fr"$\epsilon_A = {ea}, \epsilon_B = {eb}, J_0 = {Jvalue:1.3f}$", fontsize= 11)
            

            axs[ii][0].set_ylabel(r"$M^z_{A/B}$",fontsize = 14, labelpad= 0.0)  
            
            if ii ==3:
                axs[ii][1].text(20, 0.0, lbl, fontsize=12, bbox=dict(facecolor='white', alpha=1.0, edgecolor='white'))
            else:
                axs[ii][1].text(20, 0.0, lbl, fontsize=12, bbox=dict(facecolor='white', alpha=1.0, edgecolor='white'))
            
            
axs[3][0].set_xlabel(r"$t/T$", fontsize = 14, labelpad= 0.5)
axs[3][1].set_xlabel(r"$t/T$", fontsize = 14, labelpad= 0.5)

figname1 = 'reg_mz_ea_' + str(ea) + '_eb_' + str(eb) + '.svg'
figname2 = 'reg_mz_ea_' + str(ea) + '_eb_' + str(eb) + '.pdf'
plt.savefig(figname1, bbox_inches='tight', pad_inches=0.0, dpi = 300)
#plt.savefig(figname2, bbox_inches='tight', pad_inches=0.0, dpi = 600)
plt.show()
```
